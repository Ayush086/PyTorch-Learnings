```python
import torch
```

```python
x = torch.tensor(3.0, requires_grad=True) # requires_grad --> attribute indicates that we want to perform derivative on this tensor
```

```python
y = x**2
y
```

```python
y.backward() # performs the differentiation
```

```python
x.grad
```

```python
a = torch.tensor(3.0, requires_grad=True)
b = a ** 2
c = torch.sin(b)
```

```python
print(a)
print(b)
print(c)
```

```python
c.backward()
```

```python
a.grad
```
