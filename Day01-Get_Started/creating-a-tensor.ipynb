```python
import torch
torch.__version__
```

#### Creating a Tensor

```python
# using empty
t = torch.empty(2,3) # allocates the space in memory to create a tensor of given space
```

```python
type(t) # type of data structure
```

```python
# using zeros - to create a tensor where all values = 0
torch.zeros(2, 3)
```

```python
# similarly to create a tensor where values = 1
torch.ones(2, 3)
```

```python
# if we want to randomly intialize the the values, range - [0, 1]
torch.rand(2, 3)
```

```python
# now if the run the above cell multiple times then the generated values will be different every time. so if we want it to generate
# same random values every time then we must use manual_seed. Similar to what we do in ML during train test split's "random_state"
torch.manual_seed(1337)
torch.rand(1, 3)
```

```python
# for cutom tensor
torch.tensor([[1, 2, 3], [4, 5, 6]])
```

```python
# other ways

# arange
print("using arange -> ", torch.arange(0, 10, 2)) # generates a 1D tensor with provided jump

# linspace (linearly spaced)
print("using linspace -> ", torch.linspace(0, 10, 10)) # linearly spaced n-values within a given range

# eye
print("using eye -> ", torch.eye(5)) # identity matrix

# full
print('using full -> ', torch.full((3, 3), 5)) # shape of tensor with a given value at every index
```
