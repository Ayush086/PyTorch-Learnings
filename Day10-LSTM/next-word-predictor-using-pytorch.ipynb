```python
# tokenization
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# tokenize
tokens = word_tokenize(document.lower())
```

```python
# vocab building
vocabulary = {'<UNK>': 0}

for token in Counter(tokens).keys():
    if token not in vocabulary:
        vocabulary[token] = len(vocabulary)
```

```python
len(vocabulary)
```

```python
# extract sentences
input_sentences = document.split('\n')
len(input_sentences)
```

```python
def text_to_index(sentence, vocab):
    num_sentence = []
    for token in sentence:
        if token in vocabulary:
            num_sentence.append(vocabulary[token])
        else:
            num_sentence.append(vocabulary['<UNK>'])
    return num_sentence
```

```python
# indexing
input_indices = []
for sentence in input_sentences:
    input_indices.append(text_to_index(word_tokenize(sentence.lower()), vocabulary))
```

```python
len(input_indices)
```

```python
# forming training sequences
"""
eg. s = my name is yash
training sequences:
my --> name
my name --> is
my name is --> yash
"""

training_sequence = []
for sentence in input_indices:
    for i in range(1, len(sentence)):
        training_sequence.append(sentence[:i+1])
```

```python
len(training_sequence)
```

But the length of seuqences is not fixed. Therefore while training with the batches the LSTM will not be able to work parallely. Need to add padding

```python
# longest sequence
highest_len = 0
for sequence in training_sequence:
    l = len(sequence)
    if l > highest_len:
        highest_len = l

highest_len
```

```python
# zero padding (pre)
padded_training_sequence = []
for sequence in training_sequence:
    padded_training_sequence.append([0]*(highest_len - len(sequence)) + sequence)
```

```python
padded_training_sequence = torch.tensor(padded_training_sequence, dtype=torch.long)
padded_training_sequence.shape
```

```python
# training data
X = padded_training_sequence[:, :-1]
y = padded_training_sequence[: , -1]
```

```python
# Dataset
class CustomDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, index):
        return self.X[index], self.y[index]
```

```python
dataset = CustomDataset(X, y)
```

```python
# dataloader
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

#### Architecture

input layer ---> embedding layer ---> LSTM ---> output layer

```python
x = nn.Embedding(289, 100)
y = nn.LSTM(100, 150, batch_first=True)

a = dataset[0][0].unsqueeze(0)
print('Before adding to embedding layer\nshape of a: ', a.shape)

b = x(a)
print('after adding to embedding layer\n shape of b: ', b.shape)

c, d = y(b)
print('after going to LSTM layer\n shape of c: ', c.shape) # set of all intermidiate hidden state
e, f = d
print('\n shape of e: ', e.shape) # final cell state (ct)
print('\n shape of f: ', f.shape) # final hidden state

z = nn.Linear(150, 289)
g = z(f)
print('\n after going through the fully connected layer\n shape of g: ', g.shape) # probs
```

```python
class LSTMModel(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, 100)
        self.lstm = nn.LSTM(100, 150, batch_first=True)
        self.fc = nn.Linear(150, vocab_size)

    def forward(self, X):
        vector = self.embedding(X)
        intermidiate_hidden_state, (final_hidden_state, final_cell_state) = self.lstm(vector)
        output = self.fc(final_hidden_state.squeeze(0))
        return output
        
```

```python
model = LSTMModel(len(vocabulary))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else 'cpu')
device
```

```python
model.to(device)
```

```python
epochs = 50
learning_rate = 0.001

loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
# training
for epoch in range(epochs):
    total_loss = 0
    for batch_x, batch_y in dataloader:
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)

        #reset grads
        optimizer.zero_grad(x)
        # forward pass
        output = model(batch_x)
        # loss calc
        loss = loss_fn(output, batch_y)
        # back prop
        loss.backward()
        # update grads
        optimizer.step()
        total_loss = total_loss + loss.item()

    print(f"epoch: {epoch+1}, loss: {total_loss:.4f}")
```

```python
# model.to_eval()

# prediction
def predict(model, vocab, text):
    # convert to tokens
    tokenized_text = word_tokenize(text.lower())
    input_indices = text_to_index(tokenized_text, vocab)
    # padding
    padded_input = [0] * (61 - len(input_indices)) + input_indices
    # prediction
    padded_input = torch.tensor(padded_input, dtype=torch.long, device=device).unsqueeze(0)
    output = model(padded_input)
    # token to text
    max_val, idx = torch.max(output, dim=1)
    predicted_word = list(vocab.keys())[idx]

    return (text + " " + predicted_word)
```
