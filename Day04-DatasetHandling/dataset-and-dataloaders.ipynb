```python
from sklearn.datasets import make_classification
import torch
```

```python
# synthetic dataset creation
X, y = make_classification(
    n_samples=10, # no. of samples
    n_features=2, # features count
    n_informative=2, # no. of informative features
    n_redundant=0, # redundant features
    n_classes=2, # no. of classes
    random_state=43
)
```

```python
X
```

```python
# convert: np.array() --> torch.tensors
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)
```

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return self.features.shape[0] # rows count
        
    def __getitem__(self, index):
        return self.features[index], self.labels[index]
        
```

```python
dataset = CustomDataset(features=X, labels=y)
```

```python
len(dataset)
```

```python
dataset[1]
```

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)
```

```python
for batch_features, batch_labels in dataloader:
    print(batch_features)
    print(batch_labels)
    print("-"*20)
```
