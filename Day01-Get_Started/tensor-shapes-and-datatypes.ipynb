```python
import torch # type: ignore
```

```python
x = torch.tensor([[1,2,3], [4,5,6]])

x
```

```python
x.shape
```

```python
torch.empty_like(x) # copies the shape of x and allocates the space in memory to create a empty tensor
```

```python
# similarly
torch.zeros_like(x)

torch.ones_like(x)
```

```python
# also
# torch.rand_like(x) # it'll break, because rand generates the float values and x has integers in it

torch.rand_like(x, dtype=torch.float32)
```
