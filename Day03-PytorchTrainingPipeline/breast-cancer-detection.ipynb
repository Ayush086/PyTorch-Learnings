```python
df.drop(columns=['id'], inplace=True)
df.head()
```

```python
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
# scaling
scaler = StandardScaler() # mean centering

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train[0]
```

```python
# Label Encoding
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

Currently our data is in np.array() we need to convert it to tensors

```python
# np.array() -> tensors
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return self.features.shape[0]

    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
# dataset
train_dataset = CustomDataset(X_train_tensor, y_train_tensor)
test_dataset = CustomDataset(X_test_tensor, y_test_tensor)
```

```python
# dataloaders
train_data_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_data_loader = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

#### Model Definition

```python
# class firstNN():

#     # constructor
#     def __init__(self, X):
#         self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
#         self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

#     def forward_pass(self, X):
#         z = torch.matmul(X, self.weights) + self.bias
#         y_pred = torch.sigmoid(z)
#         return y_pred

#     def calculate_loss(self, y_pred, y):
#         epsilon = 1e-7
#         y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon) # to prevent log(0)

#         loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
#         return loss

import torch.nn as nn

class firstUpdatedNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        output = self.linear(features)
        output = self.sigmoid(output)
        return output
        
```

```python
# parameters
learning_rate = 0.01
epochs = 15

loss_function = nn.BCELoss()
```

```python
# training  pipeline
model = firstUpdatedNN(X_train_tensor.shape[1])

# optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)

# for epoch in range(epochs):
#     # froward prop
#     y_pred = model(X_train_tensor.float())
    
#     # calc. loss
#     loss = loss_function(y_pred, y_train_tensor.view(-1, 1).float())
    
#     optimizer.zero_grad() # clear gradients
    
#     # backward pass
#     loss.backward() # pytorch's backward prop functionality
    
#     # update params
#     optimizer.step() # parameters will get updated

#     # zero gradients - to disable the gradient accumulation
#     # model.linear.weight.grad.zero_()
#     # model.linear.bias.grad.zero_()

#     print(f'Epoch: {epoch+1}, Loss: {loss.item()}')
```

```python
for epoch in range(epochs):
    for batch_features, batch_labels in train_data_loader:
        # forward pass
        y_pred = model(batch_features.float())

        # calc. loss
        loss = loss_function(y_pred, batch_labels.view(-1, 1).float())

        # clear gradients
        optimizer.zero_grad()

        # backward pass
        loss.backward()

        # parameters update
        optimizer.step()

    print(f'Epoch: {epoch+1}, Loss: {loss.item()}')
```

```python
# evaluation using test_loader
model.eval() # set to evaluation mode
accuracy_list = []

with torch.no_grad():
    for batch_features, batch_labels in test_data_loader:
        # forward pass
        y_pred = model(batch_features.float())
        y_pred = (y_pred > 0.8).float()

        # calculate accuracy
        batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
        accuracy_list.append(batch_accuracy)

overall_accuracy = sum(accuracy_list) / len(accuracy_list)
print(f"Accuracy: {overall_accuracy: .4f}")
```
