### Scalar Operation

```python
import torch # type: ignore
x = torch.tensor([[1,2,3],[4,5,6]])
```

```python
# addition
x + 2

# subtraction
x - 2

# multiplication
x * 3

# division
x/3
```

```python
# int division
(x*100) // 3
```

### Element wise Operation

```python
a = torch.rand(2, 3)
b = torch.rand(2, 3)
```

```python
print(a)
print(b)
```

```python
# addition
a+b

# subtraction
a-b

# multiplication
a*b

# division
a/b

# power
a ** b
```

```python
## other operations

c = torch.tensor([1, -2, 3, -4])

# abs
torch.abs(c)
```

### Reduction Operation

```python
e = torch.randint(size=(2, 3), low=0, high = 10)
e
```

```python
# sum
torch.sum(e)

# sum along column
torch.sum(e, dim=0)

# sum along rows
torch.sum(e, dim=1)
```

```python
# mean
e = e.to(torch.float32)

torch.mean(e) # mean is possible only for float dtype

torch.mean(e, dim=0) # along col
torch.mean(e, dim=1) # along row
```

```python
# median
torch.median(e)

# same logic follows to perform operation along row/col

# max and min
torch.max(e)
torch.min(e)
```

```python
# product
torch.prod(e)

# st. deviation
torch.std(e)

# variance
torch.var(e)
```

```python
# argmax
torch.argmax(e) # gives the position of largest element

# argmin
torch.argmin(e) # gives the position of smallest element
```

### matrix operation

```python
f = torch.randint(size=(2, 3), low=0, high=10)
g= torch.randint(size=(3, 2), low=0, high=10)

print(f)
print(g)
```

```python
# matrix multiplication
torch.matmul(f, g)
```

```python
# dot product

v1 = torch.tensor([1, 2])
v2 = torch.tensor([3, 4])

torch.dot(v1, v2)
```

```python
# transpose
torch.transpose(g, 0, 1)
```

```python
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
# determinent
torch.det(h)

# inverse
torch.inverse(h)
```
