### Scalar Operation

```python
import torch # type: ignore
x = torch.tensor([[1,2,3],[4,5,6]])
```

```python
# addition
x + 2

# subtraction
x - 2

# multiplication
x * 3

# division
x/3
```

```python
# int division
(x*100) // 3
```

### Element wise Operation

```python
a = torch.rand(2, 3)
b = torch.rand(2, 3)
```

```python
print(a)
print(b)
```

```python
# addition
a+b

# subtraction
a-b

# multiplication
a*b

# division
a/b

# power
a ** b
```

```python
## other operations

c = torch.tensor([1, -2, 3, -4])

# abs
torch.abs(c)
```
